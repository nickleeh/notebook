# Primes numbers
(from prime list file)

## Extract txt file from zip archieve

```python
import zipfile
with zipfile.ZipFile('lib/primes1.zip', 'r') as z:
    primes_txt = z.extract('primes1.txt')
```

## Read txt file as Pandas DataFrame

```python
import pandas as pd
df = pd.read_fwf(primes_txt, 
                 header=1,
                 usecols=[0, 1, 2, 3, 4, 5, 6, 7], 
                 widths=[10,10,10,10,10,10,10,10,10,10])
df
```

## Get values from DataFrame (Result: array-like object)

```python
dv = df.values
dv
```

## Get prime numbers into list, dict, tuples

```python
primes_list = []
for l in dv:
    for x in l:
        primes_list.append(x)
```

```python
from collections import defaultdict

primes_dict = defaultdict(int)

for k,v in enumerate(primes_list):
    primes_dict[k]=v
```

```python
primes_tuple = tuple(primes_list)
```

## Write the prime numbers to a json file

```python
primes_list_normal = []

for x in primes_list:
    primes_list_normal.append(int(x))
```

```python
import simplejson as json

with open('primes.json', 'w') as f:
    json.dump(primes_list_normal, f)
```

## Write prime numbers to sqlite3 database

```python
primes_dict_normal = dict(primes_dict)

# Make a 1-indexed id record:
records = [(k+1, int(v)) for k, v in primes_dict_normal.items()] 
```

```python
import sqlite3
conn = sqlite3.connect('lib/primes.db')
c = conn.cursor()

c.executemany('insert into primes_tbl values (?, ?)', records)
conn.commit()
conn.close()
```

```python
from collections import Iterable
```

```python
def flatten(items, ignore_types=(str, bytes)):
    for x in items:
        if isinstance(x, Iterable) and not isinstance(x, ignore_types):
            yield from flatten(x)
        else:
            yield x 
```

```python
data=(90,(56,(33,(343,(34,(56,(32,(73,(54,(23,(34,(9,(4,(1,(2,(3,(4,0)))))))))))))))))
```

```python
%timeit print(sum(flatten(data)))
```

```python
sum(flatten(data))
```
